$h^{2}$ is  $x^{2}$


```python
import numpy as np
import matplotlib.pyplot as plt

# change these parameters as you need 
x_range = np.arange(0, 0.6, 0.1)
nmax = 10

An = lambda x, n, H, L: np.sin((np.pi * n * x)/ L) * ((32 * H * np.sin((3*np.pi*n)/4))/ 3*np.pi**2*n**2)
store =[]
H = 0.2 # enter actual value here
L = 0.6 # enter actual value here
for x in x_range:
    temp = 0
    for n in range(1, nmax+1):
        temp += An(x, n, H, L)
    store.append(temp)

plt.plot(x_range, store)
```
